```python
# https://stackoverflow.com/questions/16592222/matplotlib-group-boxplots

#-------------A: method-wilcoxon----------------
# gesture_distinction (m1,m2) (m1,m3) (m2,m3)



data1 = file_method_grouping.loc[file_method_grouping['plot'] == 'gesture_distinction']
data1.score = data1.score * 7 /5
fig1 = plt.figure(figsize = (10,7))
plt.ylim(3, 8)
sns.set(style="ticks", palette='Set3')
sns.set_style("whitegrid")
sns.boxplot(x='pair', y='score', hue='method', data=data1)
plt.legend(loc='lower center', title="Method", fontsize= 14)
plt.title('gesture_distinction',fontdict = {'fontsize' : 20})
```

```python
# # target_distinction (m1,m2) (m1,m3) (m2,m3)
data2 = file_method_grouping.loc[file_method_grouping['plot'] == 'target_distinction']
fig2 = plt.figure(figsize = (10,7))
plt.ylim(0, 6)
sns.set(style="ticks", palette='Set3')
sns.set_style("whitegrid")
sns.boxplot(x='pair', y='score', hue='method', data=data2)
plt.legend(loc='lower center', title="Method", fontsize= 14)
plt.title('target_distinction',fontdict = {'fontsize' : 20})
```

```python
# # naturalness (m1,m2) (m1,m3) (m2,m3)
data3 = file_method_grouping.loc[file_method_grouping['plot'] == 'naturalness']
data3.score = data3.score * 7 /5
fig3 = plt.figure(figsize = (10,7))
plt.ylim(4, 7.5)
sns.set(style="ticks", palette='Set3')
sns.set_style("whitegrid")
sns.boxplot(x='pair', y='score', hue='method', data=data3)
plt.legend(loc='lower center', title="Method", fontsize= 14)
plt.title('naturalness',fontdict = {'fontsize' : 20})
```

```python
#------------B: group-friedman-----------------
# gesture (g1, g2), target (g1, g2), naturalness (g1, g2)
data = pd.read_csv("data/data_group_grouping_friedman.csv",na_values=" NaN")
data = data.dropna(how='all', axis=1)


fig = plt.figure(figsize = (10,7))
plt.ylim(0, 6)
sns.set(style="ticks", palette='pastel')
sns.set_style("whitegrid")
sns.boxplot(x='Question', y='Likert Score (1-5)', hue='group', data=data)
plt.legend(loc='lower center', title="gesture sequence", fontsize= 14)
plt.title('Single Vs Sequential animation',fontdict = {'fontsize' : 20})
```

```python
#------------C: group-wilcoxon-----------------
# gesture (m1)(g1,g2) (m2)(g1,g2) (m3)(g1,g2)
# target (m1)(g1,g2)* (m2)(g1,g2) (m3)(g1,g2)
# naturalness (m1)(g1,g2) (m2)(g1,g2) (m3)(g1,g2)
data = pd.read_csv("data/data_group_grouping_wilcoxon.csv",na_values=" NaN")
data = data.dropna(how='all', axis=1)

data1 = data.loc[data['Question'] == 'gesture_distinction']
data2 = data.loc[data['Question'] == 'target_distinction']
data3 = data.loc[data['Question'] == 'naturalness']

```

```python
fig = plt.figure(figsize = (10,7))
plt.ylim(0, 6)
sns.set(style="ticks", palette='Set3')
sns.set_style("whitegrid")
sns.boxplot(x='Method', y='Likert Score (1-5)', hue='group', data=data1)
plt.legend(loc='lower center', title="gesture sequence", fontsize= 14)
plt.title('Gesture Distinction (Single Vs Sequential animation)',fontdict = {'fontsize' : 20})
```
